```python
#markers graph
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([9, 3, 6, 2])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["bat", "ball", "gloves", "inner thypad"]
mycolors = ["brown", "red", "orange", "blue"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
#pie chart with shadow
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["bat", "Ball", "gloves", "shoe"]
myexplode = [0.2, 0, 0, 0.3]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```
